```python
from torchvision import datasets, transforms
import torch
from torch.utils.data import DataLoader, WeightedRandomSampler
import numpy as np

# 1) Görüntü dönüştürme işlemleri
transform_train = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5,0.5,0.5], [0.5,0.5,0.5])
])

# 2) ImageFolder ile veri kümelerini oluşturma
train_dataset = datasets.ImageFolder(
    root="/content/frames/frames/train",
    transform=transform_train
)
val_dataset = datasets.ImageFolder(
    root="/content/frames/frames/val",
    transform=transform_train
)
test_dataset = datasets.ImageFolder(
    root="/content/frames/frames/test",
    transform=transform_train
)

print("Train samples:", len(train_dataset),
      "Val samples:", len(val_dataset),
      "Test samples:", len(test_dataset))

# 3) Her sınıfın (REAL=0, FAKE=1) kaç örneği olduğunu sayalım:
real_count = len([path for (path, label) in train_dataset.samples if label == 0])
fake_count = len([path for (path, label) in train_dataset.samples if label == 1])

print(f"REAL sınıfı sayısı: {real_count}, FAKE sınıfı sayısı: {fake_count}")

# 4) Sınıf ağırlıklarını belirleyelim (örneğin ters orantılı)
#    Daha yüksek 'weight' demek, modelin o sınıfa daha fazla odaklanacağı anlamına gelir
class_weights = [1.0 / real_count, 1.0 / fake_count]  # [weight_real, weight_fake]

# 5) Tüm train_dataset örnekleri için 'sample weight' hesapla
samples_weights = [class_weights[label] for (_, label) in train_dataset.samples]
samples_weights = torch.DoubleTensor(samples_weights)  # PyTorch tensör

# 6) WeightedRandomSampler tanımla
sampler = WeightedRandomSampler(
    weights=samples_weights,
    num_samples=len(samples_weights),  # Bir epoch'ta kaç örnek seçilsin
    replacement=True  # Aynı örnek tekrar seçilebilir, ama ortalama sınıf dağılımını eşitler
)

# 7) train_loader'ı artık sampler ile oluştur
train_loader = DataLoader(
    train_dataset,
    batch_size=64,
    sampler=sampler,   # shuffle yerine sampler kullanılır
    num_workers=4
)

# Val ve Test loader eskisi gibi (dengesizlik genelde train set için kritik)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False, num_workers=4)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False, num_workers=4)

# Artık train_loader, az olan sınıfa daha fazla örnek vermeye çalışacak
```

```python
from torchvision import datasets, transforms
import torch

transform_train = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize([0.5,0.5,0.5], [0.5,0.5,0.5])
])

# ImageFolder, alt klasör isimlerini otomatik etiket (REAL=0, FAKE=1) olarak kabul eder
train_dataset = datasets.ImageFolder(root="/content/frames/frames/train",
                                     transform=transform_train)
val_dataset   = datasets.ImageFolder(root="/content/frames/frames/val",
                                     transform=transform_train)
test_dataset  = datasets.ImageFolder(root="/content/frames/frames/test",
                                     transform=transform_train)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=4)
val_loader   = torch.utils.data.DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=4)
test_loader  = torch.utils.data.DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=4)

print("Train samples:", len(train_dataset), "Val samples:", len(val_dataset), "Test samples:", len(test_dataset))
```

```python
import torchvision.models as models
from torchvision.models import efficientnet_b4, EfficientNet_B4_Weights

class EfficientNetDF(nn.Module):
    def __init__(self, num_classes=2, dropout_rate=0.5):
        super().__init__()

        # EfficientNet-B4 yüklüyoruz (daha büyük model)
        self.model = efficientnet_b4(weights=EfficientNet_B4_Weights.IMAGENET1K_V1)

        # EfficientNet-B4'ün son katmanını devre dışı bırakıyoruz
        in_features = self.model.classifier[1].in_features
        self.model.classifier[1] = nn.Identity()

        # Yeni katmanlar (Dropout + Linear)
        self.dropout = nn.Dropout(p=dropout_rate)
        self.classifier = nn.Linear(in_features, num_classes)

    def forward(self, x):
        x = self.model(x)
        x = self.dropout(x)
        x = self.classifier(x)
        return x


# Modeli oluşturup CUDA'ya gönderiyoruz
model = EfficientNetDF(num_classes=2, dropout_rate=0.5).cuda()

criterion = nn.CrossEntropyLoss()

optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=1e-5)

def train_one_epoch(model, loader, optimizer, criterion):
    model.train()
    total_loss, total_acc, total_samples = 0, 0, 0

    for images, labels in loader:
        images, labels = images.cuda(), labels.cuda()

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * images.size(0)
        preds = torch.argmax(outputs, dim=1)
        total_acc += (preds == labels).sum().item()
        total_samples += images.size(0)

    avg_loss = total_loss / total_samples
    avg_acc = total_acc / total_samples
    return avg_loss, avg_acc

def eval_one_epoch(model, loader, criterion):
    model.eval()
    total_loss, total_acc, total_samples = 0, 0, 0

    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.cuda(), labels.cuda()

            outputs = model(images)
            loss = criterion(outputs, labels)

            total_loss += loss.item() * images.size(0)
            preds = torch.argmax(outputs, dim=1)
            total_acc += (preds == labels).sum().item()
            total_samples += images.size(0)

    avg_loss = total_loss / total_samples
    avg_acc = total_acc / total_samples
    return avg_loss, avg_acc
```
